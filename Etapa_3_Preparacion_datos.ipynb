**Agregacion de la columna vivos**

1 = vivos en df_nac  => Nacidos vivos 2018

-1 = no vivo en df_fet => Fetales muertos 2018

```python
df_nac['VIVOS'] = 1 #Aqui lo que hago es que se llama el dataframe de nacidos vivos 2018, crear la columna VIVOS y que todas sean iguales a 1
df_fet['VIVOS'] = -1 #Aqui lo que hago es que se llama el dataframe de fetales muertos 2018, crear la columna VIVOS y que todas sean iguales a -1
```

```python
# aqui se declara la funcion re_categorizacion esto sirve para reemplazar en Sitio de parto (la columna SIT_PARTO)
# para que los datos coincidan a la hora de adicionar los registros al otro dataframe
# Los 2 se reemplazan por 3
# Los 3 se reemplazan por 6
def re_categorizacion(row):
    try:
        if (2==row):
            return 3
        else:
            if(3==row):
                return 6
            else:    
                return row 
    except:
        return np.NaN #si hay problemas, no retorne nada

```

```python
#Aqui se aplica un filtrado del metodo re_categorizacion que se hablo en la Etapa II - Entendimiento de los datos, que datos se iban a recategorizar
df_nac['SIT_PARTO'] = df_nac['SIT_PARTO'].apply(re_categorizacion)
df_nac
```

## Unir todos los dataframes en uno solo

## renombrar columnas

renombrar las columnas del dataframe df_fet antes de hacer la adicion de datos

```python
# Renombrar columnas en el dataframe fetales
df_fet = df_fet.rename(
    columns={
        "A_DEFUN": "AREANAC",
        "SIT_DEFUN": "SIT_PARTO",
        "PESO_NAC": "PESO_NAC",
        "T_GES": "T_GES",
        "T_PARTO": "TIPO_PARTO",
        "TIPO_EMB": "MUL_PARTO",
        "EDAD_MADRE": "EDAD_MADRE",
        "SEG_SOCIAL": "SEG_SOCIAL",
         "IDPROFCER": "PROFESION",
    }
)
```

## Columnas prometedoras

```python
# Tienen que ser iguales para hacer la adicion, y con esto armo listas para filtrarlas

# Columnas prometedoras nac
prom_columns_nac = [ 'AREANAC','SIT_PARTO','PESO_NAC',
                'T_GES','TIPO_PARTO', 'MUL_PARTO',
                'EDAD_MADRE','SEG_SOCIAL','PROFESION', 'VIVOS'
]

# Columnas prometedoras fet
prom_columns_fet = [ 'AREANAC','SIT_PARTO','PESO_NAC',
                'T_GES','TIPO_PARTO', 'MUL_PARTO',
                'EDAD_MADRE','SEG_SOCIAL','PROFESION', 'VIVOS'
]
```

## Agregación de los datos (adición)

```python
#Añadir las filas correspondientes de los dataframes para unirlos y que queden en uno solo
df_unificado = df_nac[prom_columns_nac].append(df_fet[prom_columns_fet], ignore_index=True)
df_unificado
```

## Limpieza de los datos

```python
#revision de valores nulos e todos las columnas
df_unificado.isna().sum()
```

## Filtrado

```python
#Eliminar todos los valores que terminen en = 9 o 99 en uno pocos casos. Con filtrados

#En este caso no vamos a eliminar... Nosotros vamos a filtrar los valores donde no haya información
#Si termina en 9 es porque no tiene informacion
#Si termina en 99 en Edad Madre es porque no tiene informacion 
#Ya esta definido en la Etapa II

df_unificado = df_unificado[df_unificado.AREANAC!=9]
df_unificado = df_unificado[df_unificado.SIT_PARTO!=9]
df_unificado = df_unificado[df_unificado.PESO_NAC!=9]
df_unificado = df_unificado[df_unificado.T_GES!=9]
df_unificado = df_unificado[df_unificado.TIPO_PARTO!=9]
df_unificado = df_unificado[df_unificado.MUL_PARTO!=9]
df_unificado = df_unificado[df_unificado.EDAD_MADRE!=99] #este caso es especial, solo los 99 le eliminan
df_unificado = df_unificado[df_unificado.SEG_SOCIAL!=9]
df_unificado = df_unificado[df_unificado.PROFESION!=9]
df_unificado = df_unificado[df_unificado.VIVOS!=9]
df_unificado
```

## Recategorizacion en dummy y ordinal

```python
# Creamos un dataframe para separar el df_unificado y trabajar con un nuevo dataframe
# Para no generar conflictos, ya que podria sobreescribir la informacion y de pronto se necesita por se renombra una nueva variable dataframe que tenga en contenido de df_unificado
df_categorizado = df_unificado
df_categorizado
```

```python
df_categorizado.describe() #Aqui esta todo el dataframe adicionado, es decir los dos estan en uno solo.
#Lo que hacemnmos aqui es un analisis para saber que columnas posiblemente se puede obtener un valor de peso, eso es lo que pudimos determinar
```

```python
#Generar columnas dummy AREANAC y SEG_SOCIAL y las ordinales se quedan tal cual como estan 
df_categorizado = pd.concat([df_categorizado,pd.get_dummies(df_categorizado['AREANAC'],prefix='AREANAC')],axis=1)
df_categorizado = pd.concat([df_categorizado,pd.get_dummies(df_categorizado['SEG_SOCIAL'],prefix='SEG_SOCIAL')],axis=1)

df_categorizado 
```

```python
#Borrar columnas no usadas
#Aqui elimino por columnas, estas dos columnas que no usare
df_categorizado.drop('AREANAC',axis=1,inplace=True)
df_categorizado.drop('SEG_SOCIAL',axis=1,inplace=True)
df_categorizado 
```
